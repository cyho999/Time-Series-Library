```python
# read a .npy file and plot the data
import numpy as np
import matplotlib.pyplot as plt

a = np.load("pred.npy")
b = np.load("true.npy")
b.shape

for i in range(12):
    plt.plot(b[:,i,:], label="true")
    plt.plot(a[:,i,:], label="pred")
    plt.legend()
    plt.title(f"leadtime: {i}")
    plt.show()
    plt.clf()
```
